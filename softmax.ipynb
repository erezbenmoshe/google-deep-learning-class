# Illustrates softmax function behavior
### Maps logits (scores) to probability vectors that sum to 1

```python
import numpy as np
```

```python
def softmax(x):
    return np.exp(x) / sum(np.exp(x))
```

```python
x = np.arange(-2.0, 6.0, 0.1)
```

```python
x
```

```python
np.ones_like(x)
```

```python
scores = np.vstack([x, np.ones_like(x), 0.2*np.ones_like(x)])
```

```python
scores.shape
```

```python
scores
```

```python
softmax(scores)
```

```python
sum(softmax(scores))   # sums column wide (adds to 1) 
```

```python
s = [3.0, 1.0, 0.2]
```

```python
softmax(s)
```
